```python
import requests

# 1. Your API Key
API_KEY = "sk_xwuv0b6h_v5ZzxKbbVTvgdyYt4HePMaL6"

# 2. Your Status Response (this represents the object you received)
status_response = job.wait_until_complete()  # Or the JobStatusV1Response object you have

# 3. Extract Job ID and File ID (from the outputs section)
job_id = status_response.job_id

# In your response, the transcript file is named '0.json' 
# It is located inside job_details -> outputs list
file_id = status_response.job_details[0].outputs[0].file_name 

# 4. Create the Payload
payload = {
    "job_id": job_id,
    "files": [file_id]  # Must be a list of strings
}

# 5. Define Headers
headers = {
    "api-subscription-key": API_KEY,
    "Content-Type": "application/json"
}

# 6. Send the Request to get Presigned URLs
url = "https://api.sarvam.ai/speech-to-text/job/v1/download-files"
response = requests.post(url, json=payload, headers=headers)

if response.status_code == 200:
    # 7. Get the download URL from the response
    download_urls = response.json().get("urls", [])
    if download_urls:
        presigned_url = download_urls[0]
        
        # 8. Fetch the ACTUAL JSON data from the link
        transcript_json = requests.get(presigned_url).json()
        
        print("--- Success! Transcript Found ---")
        print(transcript_json.get("transcript"))
else:
    print(f"Error {response.status_code}: {response.text}")
```
